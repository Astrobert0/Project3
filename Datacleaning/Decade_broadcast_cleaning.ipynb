```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import re

path = r'C:\Users\Sebas!\Documents\GitHub\Project3\decade_broadcast.csv'
data = pd.read_csv(path)
data = pd.DataFrame(data)

data.index = ["Views"]
data= data.transpose()#in billions
location = []
year = []
for word in data.index: 
    number = re.findall(r'\d{4}', word)
    city = re.findall(r'[a-zA-Z\s]+', word) #this findall creates a list with 9 values of list with 1 value
                                            #[[a],[b]]...]
    location.append(city[0].strip())
    year.append(number[0])
data['Location'] = location #add new columns
data['Year'] = year
data.sort_values(by = "Year", ascending=True, inplace =True)
#data['Year'].astype('int32')

data.to_csv(r'C:\Users\Sebas!\Documents\GitHub\Project3\decade_broadcast.csv')
```

```python
#making the chart
plt.plot(data['Year'], data['Views'], 'b-')
plt.ylabel("Views in billions")
plt.xlabel("Years")
plt.title("Olympics total broadcast")
plt.show()
#data.plot(kind= 'line', x = 'Year', y= 'Views')
```
