```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

decade_broadcast = pd.read_csv(r'C:\Users\Sebas!\Documents\GitHub\Project3\decade_broadcast.csv')
decade_broadcast.rename(columns ={'Unnamed: 0':'Event'}, inplace = True)

TV_subs = pd.read_csv(r'C:\Users\Sebas!\Documents\GitHub\Project3\TV_subscribers_to_2015.csv')
TV_subs.drop(columns = 'Telephone company TV subscribers', inplace = True)
TV_subs.rename(columns = {'Cable TV subscribers (In millions)':'Cable TV subscribers'}, inplace = True)
```

```python
#TV Subscriptions
df = decade_broadcast[['Year','Views']]
df['Views'] = df['Views'] * 10
compare_chart = TV_subs.merge(df,how= 'right',left_on='Year', right_on = 'Year')

x_data = compare_chart['Year']
y_data1 = compare_chart['Views']
y_data2 = compare_chart['Cable TV subscribers']

plt.plot(x_data, y_data1, color = 'r', linestyle = '--',marker= '.', label = 'Olympics broadcast')
plt.plot(x_data, y_data2, color = 'b',marker = '.', label = 'US TV subscriptions')
plt.title('US TV subscriptions VS Olympics broadcast')
plt.legend()
plt.savefig('BroadcastVStv_Subscription.png')
```

```python
#Olympics broadcast
y_axis2 = decade_broadcast['Views']
x_axis2 = decade_broadcast['Event']

sns.set(font_scale = 2 ,rc={'figure.figsize':(32,13)})
sns.barplot( x = x_axis2, y = y_axis2, data= decade_broadcast)
plt.ylabel('Views (in billions)')
plt.title('Global broadcast')
plt.savefig('Decade_Broadcast.png')
```
